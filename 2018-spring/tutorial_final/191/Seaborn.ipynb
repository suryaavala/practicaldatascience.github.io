## Matplotlib vs Seaborn

Seaborn is actually based on matplotlib with more advanced API encapsulation, which makes visualizing data easier. In most cases, seaborn can be used to make an attractive chart, and matplotlib can be used to create more charts with different features. Seaborn should be considered a supplement to matplotlib but not a replacement.


## Style Setting

Seaborn divides the matplotlib parameters into two separate groups. The first group sets the aesthetic style, and the second group is a different measure element so that it can be easily added to the code.

The interfaces that operate on these parameters are two pairs of functions. To control the style, use the functions `axes_style()` and `set_style()`. To extend the plot, use the `plotting_context()` and `set_context()` functions. In both cases, the first function returns a parameter dictionary, and the second function sets the matplotlib default properties.

`set_style()` provides 5 different default styles:  
    
    
    darkgrid 
    
    whitegrid 
    
    dark  
    
    white
    
    ticks  
    
    
the default style is darkgrid

`axes_style()` can set or return the current axes parameters.

You can use `despine()`to control the bounds of the frame. 


`despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)` 




```python
sns.set_style("darkgrid")  
plt.plot(np.arange(200))  
plt.show() 

sns.set_style("whitegrid")  
plt.plot(np.arange(200))  
plt.show() 

sns.set_style("white")  
plt.plot(np.arange(200))  
plt.show() 

sns.set_style("dark")  
plt.plot(np.arange(200))  
plt.show() 

sns.set_style("ticks")  
plt.plot(np.arange(200))  
plt.show() 

sns.set_style("white")  
plt.plot(np.arange(200))  
sns.despine()
plt.show()

sns.set_style("white")  
plt.plot(np.arange(200))  
sns.despine(left = True, bottom = True)
plt.show()

print(sns.axes_style())
```

## Plotting

In his chapter, I am going to introduce how to plot different charts that frequently been using in data visualization.


#### Histograms 

We will use `distplot()` to construct the histogram in Seaborn, however, Seaborn not only allow us to use the method to draw histograms, but also adding many other features into the chart, which can concentrate the informations.

`distplot(a, bins=None, hist=True, kde=True, rug=False, fit=None, hist_kws=None, kde_kws=None, rug_kws=None, fit_kws=None, color=None, vertical=False, norm_hist=False, axlabel=None, label=None, ax=None)`


`hist` controls the display of the histogram chart

`ked` controls the display of the kernel density estimate.

`rug` controls the display of the rugplot on the support axis.


```python
sns.distplot(myDataFrame["Power_Generation(Billion kWh)"], bins = 20, kde=False, rug=False)
plt.show()
```

```python
sns.distplot(myDataFrame["Power_Generation(Billion kWh)"], bins = 20, kde=True, rug=False)
plt.show()
```

```python
sns.distplot(myDataFrame["Power_Generation(Billion kWh)"], bins = 20, kde=True, rug=True,color = "r")
plt.show()
```

You can call `kdeplot()` if you just want to plot the kernal density estimation

```python
sns.kdeplot(myDataFrame["Power_Generation(Billion kWh)"],color = "g")
plt.show()
```

#### Boxplot

If you are dealing with multiple columns of same series of data, `boxplot()` will be a good tool for you.
As mentioned in the official documentation:"A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range."

`seaborn.boxplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, width=0.8, dodge=True, fliersize=5, linewidth=None, whis=1.5, notch=False, ax=None)`


```python
dotsData = sns.load_dataset("dots")
sns.boxplot(x="coherence",y="firing_rate",hue ="choice", data = dotsData,palette = sns.color_palette("coolwarm"))
plt.show()
```

#### Stripplot &  Swarmplot

The essence of `stripplot()` and `swarmplot()` is using the quantitative characteristics variants to make to scatterplot. If some of the data overlap with each other, we can set the parameter `jitter = True` to provide us a better visualization quality.


```python

with sns.axes_style("whitegrid"): 
    sns.stripplot(x='coherence',y='firing_rate',data=dotsData,jitter = True) 

plt.show()
```

However, if there are too many data overlap with each other, another solution is using the `swarmplot()`

```python

with sns.axes_style("whitegrid"): 
    sns.swarmplot(x='coherence',y='firing_rate',data=dotsData) 

plt.show()
```

#### Pointplot

If we need a scatter plot or point plot,we can use the `pointplot()`:

Also you can also draw the Hexbin chart which will propriate to implement on a large set of data to show the relation between bivariances.

## Linear Regression

While we are dealing with dataset, the set may contain many different invariants and variants. The goal of analysis is to connect some of the variants to reveal a deeper layer of information. Say we've already have a set of data, and we want to excavate the variants' relation and predict the output in the future based on the input we will feed in. It will be pretty intuitive for us to use the scikit-learn library and plot the chart after the calculation and fit-in. However, the Seaborn library provides us a simple and fast way to come up with a prediction and get the data visualization in the same time. What Seaborn will provide is a fast and convenient way to get the prediction curve, in another word, if we want a solid and robust or deal with a complicate problem, using other professional library will make more sense. 




### Drawing the regression curve

Use the two main functions in Seaborn to visualize the linear relationship determined by regression. These functions `regplot()` and `lmplot()` are closely related and share most of the core functionality. However, it is important to understand their different ways so that you can quickly choose the right tool for your specific job.

In the simplest call, two functions plot a scatterplot of two variables x and y, then fit the regression model y~x and plot the regression line and 95% confidence intervals for the regression line.

`seaborn.regplot(x, y, data=None, x_estimator=None, x_bins=None, x_ci='ci', scatter=True, fit_reg=True, ci=95, n_boot=1000, units=None, order=1, logistic=False, lowess=False, robust=False, logx=False, x_partial=None, y_partial=None, truncate=False, dropna=True, x_jitter=None, y_jitter=None, label=None, color=None, marker='o', scatter_kws=None, line_kws=None, ax=None)`


`seaborn.lmplot(x, y, data, hue=None, col=None, row=None, palette=None, col_wrap=None, size=5, aspect=1, markers='o', sharex=True, sharey=True, hue_order=None, col_order=None, row_order=None, legend=True, legend_out=True, x_estimator=None, x_bins=None, x_ci='ci', scatter=True, fit_reg=True, ci=95, n_boot=1000, units=None, order=1, logistic=False, lowess=False, robust=False, logx=False, x_partial=None, y_partial=None, truncate=False, x_jitter=None, y_jitter=None, scatter_kws=None, line_kws=None)`

```python
sns.regplot(x="Power_Generation(Billion kWh)", y="Coal-Fired Power Generation", data=myDataFrame)
plt.show()
```

```python
sns.lmplot(x="Power_Generation(Billion kWh)", y="Coal-Fired Power Generation", data=myDataFrame)
plt.show()
```

`regplot()` accepts different kinds of x and y variants, but `lmplot()`requires the user to feed in the data as well. 
for example :

The plot about can be generate without feeding in the dataset:


```python
sns.set_style("dark")
sns.regplot(x= myDataFrame["Power_Generation(Billion kWh)"], y = myDataFrame["Coal-Fired Power Generation"])
plt.show()
```

However, if the data doesn't have a linear relationship, the Seaborn library also provide nonlinear models to fit in the data. We can use the `order=` parameter.

```python
sns.regplot(x="Nuclear Power Generation", y="Coal-Fired Power Generation", data=myDataFrame,order = 2)
plt.show()
```

`residplot()` is a useful tool for checking if a simple regression model fits a data set. It fits and removes a simple linear regression and then plots the residual value of each observation. Ideally, these values should be randomly scattered around y = 0

```python
sns.residplot(x="Nuclear Power Generation", y="Coal-Fired Power Generation", data=myDataFrame)
plt.show()
```

Similarly, the regression plot can also be jointed with histograms or kde by the `jointplot()`

```python
sns.jointplot(x="Power_Generation(Billion kWh)", y="Coal-Fired Power Generation", data=myDataFrame,kind = "reg" )
plt.show()
```
